```python
from YearPlotter import YearPlotter
def plot_stt_mea(fig,ax,stt='USC00486440', mea='PRCP'):
    Query="SELECT * FROM weather\n\tWHERE measurement='%s' and station='%s'"%(mea,stt)
    #print Query
    df1 = sqlContext.sql(Query)
    #print df1.count(),'rows'
    rows=df1.rdd.map(lambda row:unpackArray(row['vector'],np.float16)).collect()
    T=np.vstack(rows)
    if mea[0] == 'T':
        T=T/10.  # scaling to make the temperature be in centingrates
    #shape(T)
    YP=YearPlotter()
    YP.plot(T[:3,:].transpose(),fig,ax,title=mea)
    #title('station: %s measurement: %s'%(stt, mea))
def plot_stt_meas(stt='USC00486440'):
    meas = ['TMIN', 'TMAX', 'TOBS','SNOW', 'SNWD', 'PRCP']
    k=0
    fig,axes=plt.subplots(1,3,figsize=(20,4))
    for i in range(3):
        axis = axes[i]
        plot_stt_mea(fig,axis,stt,meas[k])
        k+=1
    fig,axes=plt.subplots(1,3,figsize=(20,4))
    for i in range(3):
        axis = axes[i]
        plot_stt_mea(fig,axis,stt,meas[k])
        k+=1
```

```python
YP = YearPlotter()
def plot_func(func):
    meas = ['TMIN', 'TMAX', 'TOBS', 'SNOW', 'SNWD', 'PRCP']
    k=0
    fig,axes=plt.subplots(1,3,figsize=(20,4))
    for i in range(3):
        axis = axes[i]
        func(fig,axis,meas[k])
        k+=1
    fig.savefig('figs/TMIN_TMAX_TOBS_%s.png'%(func.__name__[5:]))
    fig,axes=plt.subplots(1,3,figsize=(20,4))
    for i in range(3):
        axis = axes[i]
        func(fig,axis,meas[k])
        k+=1
    fig.savefig('figs/SNOW_SNWD_PRCP_%s.png'%(func.__name__[5:]))

def plot_valid(fig,axis,m):
    valid_m=STAT[m]['NE']
    YP.plot(valid_m,fig,axis,title='valid-counts '+m)
    
def plot_mean_std(fig,axis,m):
    mean=STAT[m]['Mean']
    std=np.sqrt(STAT[m]['Var'])
    graphs=np.vstack([mean-std,mean,mean+std]).transpose()
    YP.plot(graphs,fig,axis,title='Mean+-std   '+m)
    
def plot_eigen(fig,axis,m):
    EV=STAT[m]['eigvec']
    YP.plot(EV[:,:3],fig,axis,title='Top Eigenvectors '+m)
```
