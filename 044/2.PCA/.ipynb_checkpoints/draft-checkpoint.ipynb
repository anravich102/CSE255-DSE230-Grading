### 2 Discriminative Learning

Question 1: Perform a 80-20 split using train test split on the data to obtain the train and the test data (random state=3). Use Logistic Regression to classify the wines according to their cultivators. Tune parameters ’penalty’ and ’C’ using GridSearchCV implementation. Report the accuracy on test data.

```python
# wine_original.csv for this question
# Read the csv as pandas data frame
wine_original_df = pd.read_csv('wine_original.csv')
#print wine_original_df.shape
#print wine_original_df.head(2)

# Create input and output data from given data file
X = wine_original_df.iloc[:,1:].values
Y = wine_original_df['class'].values

# Data set split into train and test as per given random state
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=3)

# Penalty specify norm used in penalization
# C is inverse of regularization strength and smaller values of it specify stronger regularization
# Increase 'C' values was resulting in picking larger 'C' as best parameter and poor test accuracy
parameters = { 'penalty': ['l1','l2'], 
               'C':[0.1, 0.5, 1, 2, 3, 4, 5, 10]} 

# Logistic regression classifier
logreg = LogisticRegression()
# this n_jobs = -1 makes algorithm to use all cores of machine
logreg_clf = GridSearchCV(logreg, parameters, verbose=True, n_jobs=-1) 
logreg_clf.fit(X_train, Y_train)
Y_pred = logreg_clf.predict(X_test)

# test accuracy
test_accuracy = round(accuracy_score(Y_test,Y_pred),5)

print ('Selected Parameters are ', logreg_clf.best_params_)
print ('Test Accuracy = ' + str(test_accuracy))
```

#### Solution for Question1: 
- After tuning parameters 'penalty' and 'C', selected parameters are l1 norm for penalty and 1 for C
- For these chosen parameters, Test Accuracy is 0.888888888889

#### 3.1 Text data processing

```python
# newsgroups data is loaded from sklearn data sets with below filters
  # 1. Headers, footers and quotes are removed
  # 2. After Data loaded is only for few categories specified in problem
from sklearn.datasets import fetch_20newsgroups
newsgroups_train = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'),categories= ('alt.atheism', 'comp.graphics', 'sci.space', 'talk.politics.mideast'))
newsgroups_test = fetch_20newsgroups(subset='test', remove=('headers', 'footers', 'quotes'),categories= ('alt.atheism', 'comp.graphics', 'sci.space', 'talk.politics.mideast'))

# Read the training data and labels from the data set
X_train_raw = newsgroups_train.data
Y_train = newsgroups_train.target
X_test_raw  = newsgroups_test.data
Y_test  = newsgroups_test.target
```

```python
# - Convert all the words in the text to lower case. 
# - Remove stop words
# - Select only top 2000 features
# All above can be acheived with TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english',lowercase=True,max_features=2000)
X_train = vectorizer.fit_transform(X_train_raw)
X_test = vectorizer.transform(X_test_raw)
#print(X_train.shape)
#print(X_test.shape)
```

Question 2: After obtaining the tf-idf vectors for train and test data, use the perceptron model (no penalty) to train on the training vectors and compute the accuracy on the test vectors.

Question 6: Perform 80-20 split of the training data to obtain validation data using train test split (random state=10). Use this validation data to tune the cost parameter ’C’ for values 0.01,0.1,1,10,100. Select the best value compute the accuracy for the test data. Report the validation and test accuracies. Note: Use full data of 2000 vectors here.

```python
# Since tf-idf vectors on basic data is specified for 2000 features, computing accuracy for that 2000 features
vectorizer = TfidfVectorizer(stop_words='english',lowercase=True,max_features=2000)
X_train = vectorizer.fit_transform(X_train_raw)
X_test  = vectorizer.transform(X_test_raw)

# Data set split into train and test as per given random state
X_train_new, X_valid, Y_train_new, Y_valid = train_test_split(X_train, Y_train, test_size = 0.2, random_state=10)
```

```python
# C is coefficient to control margin and error
C = [0.01, 0.1, 1, 10, 100]

opt_C = 0
acc_opt_C = 0

for c in C:
    SVM_clf = SVC(C = c, kernel='linear') # considering linear kernel here
    SVM_clf.fit(X_train_new, Y_train_new)
    Y_pred = SVM_clf.predict(X_valid)
    acc_per_C = accuracy_score(Y_valid, Y_pred)
    print('Validation accuracy for C =' + str(c) + '\tis ' + str(acc_per_C))
    
    #checking if accuracy for that C is greater than optimal C
    # if so assign this greater value and C values to optimal accuracy and optimal C respectively
    if (acc_per_C > acc_opt_C):
        opt_C = c
        acc_opt_C = acc_per_C
        
print('Best validation accuracy is ' + str(acc_opt_C) + ' for C value =' + str(opt_C))
```

```python
# Test accuracy is calculated using concatenated training data i.e. original X_train and Y_train
SVM_clf = SVC(C = c, kernel='linear') # considering linear kernel here
SVM_clf.fit(X_train, Y_train)
Y_pred = SVM_clf.predict(X_test)
acc_score = accuracy_score(Y_test, Y_pred)
print('Test accuracy for C = ' + str(opt_C) + '\tis ' + str(acc_score))
```

b) Let K1 be the ’cosine similarity’ and K2 be ’Laplacian Kernel’. Using K as kernel, train a SVM model to tune the value of α (upto one decimal) and report the accuracy on the test data using the selected parameter.

```python
# SVM model need to be tuned for different 
# Since this customer Kernel is only valid kernel in this range and given tune upto one decimal point, 
# 'alpha' is tuned from 0 to 1 with step size of .1

# Step1: Tune the model for this alpha range from [0,1]
alpha_val = np.arange(0, 1.01, 0.1)

opt_alpha = 0
acc_opt_alpha = 0
for alpha in alpha_val:
    #print alpha
    # From documentation of sklearn, if Y is not specified, it is pair wise similarities between itself
    X_train_K1 = (alpha * cosine_similarity(X_train_new)) 
    X_train_K2 = ((1-alpha)* laplacian_kernel(X_train_new))
    X_train_K = (X_train_K1+X_train_K2)
    
    X_valid_K1 = (alpha * cosine_similarity(X_valid,X_train_new)) 
    X_valid_K2 = (1-alpha) * laplacian_kernel(X_valid,X_train_new)
    X_valid_K = (X_valid_K1+X_valid_K2)
    
    
    SVM_clf = SVC(kernel='precomputed') # considering precomputed or customer kernel
    SVM_clf.fit(X_train_K, Y_train_new)
    Y_pred = SVM_clf.predict(X_valid_K)
    acc_per_alpha = accuracy_score(Y_valid, Y_pred)
    print('Validation accuracy for alpha (α) = ' + str(alpha) + '\tis ' + str(acc_per_alpha))
    
    #checking if accuracy for that alpha is greater than optimal alpha
    # if so assign this greater value and alpha values to optimal accuracy and optimal alpha respectively
    if (acc_per_alpha > acc_opt_alpha):
        opt_alpha  = alpha
        acc_opt_alpha = acc_per_alpha # update optimal alpha with latest values
        
print('Highest validation accuracy is ' + str(acc_opt_alpha) + ' for α value =' + str(opt_alpha))


```

```python
X_train_K1 = (opt_alpha * cosine_similarity(X_train_new)) 
X_train_K2 = ((1-opt_alpha)* laplacian_kernel(X_train_new))
X_train_K = (X_train_K1+X_train_K2)
    
X_test_K1 = (opt_alpha * cosine_similarity(X_test,X_train_new)) 
X_test_K2 = ((1-opt_alpha) * laplacian_kernel(X_test,X_train_new))
X_test_K = (X_test_K1+X_test_K2)


SVM_clf = SVC(kernel='precomputed')  # considering precomputed or customer kernel
SVM_clf.fit(X_train_K, Y_train_new)
Y_pred = SVM_clf.predict(X_test_K)

acc_per_alpha = accuracy_score(Y_test, Y_pred)
print('Test accuracy for alpha (α) = ' + str(opt_alpha) + '\tis ' + str(acc_per_alpha))
    
```
