```python
import pandas as pd
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
import statsmodels.api as sm2
```

## !pwd

```python
df = pd.read_csv("LinearRegression.csv")
```

```python
result = sm.ols(formula="AvgTemp ~ Precipitation + SnowfallDep", data=df).fit()
```

```python
print result.params
```

```python
print result.summary()
```

```python
plt.show()
```

```python
sm2.graphics.plot_fit(result,"Precipitation")
```

```python
sm2.graphics.plot_fit(result,"SnowfallDep")
```
