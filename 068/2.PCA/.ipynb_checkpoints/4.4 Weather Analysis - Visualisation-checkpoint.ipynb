### Distribution of missing observations
The distribution of missing observations is not uniform throughout the year. We visualize it below.

```python
def plot_pair(pair,func,ylabels=None,tenths=None):
    j=0
    fig,X=subplots(1,2,figsize=(16,6))
    axes=X.reshape(2)
    for m in pair:
        axis = axes[j]
        ylabel=None
        if ylabels is not None:
            ylabel=ylabels[j]
        tenth=None
        if tenths is not None:
            tenth = tenths[j]
        j+=1
        func(m,fig,axis,ylabel,tenth)
        
def plot_valid(m,fig,axis,ylabel=None,tenth=None):
    valid_m=STAT[m]['NE']
    if ylabel is not None:
        axis.set_ylabel(ylabel)
    YP.plot(valid_m,fig,axis,title='valid-counts '+m)
    
```

```python
plot_pair(['TMIN','TMAX'],plot_valid)
```

```python
plot_pair(['TOBS','PRCP'],plot_valid)
```

```python
plot_pair(['SNOW', 'SNWD'],plot_valid)
```

### Plots of mean and std of observations

```python
def plot_mean_std(m,fig,axis,ylabel=None,tenth=None):
    if tenth is True:
        mean=STAT[m]['Mean'] / 10.0
        std=np.sqrt(STAT[m]['Var']) / 10.0
    else:
        mean=STAT[m]['Mean']
        std=np.sqrt(STAT[m]['Var'])
    graphs=np.vstack([mean+std,mean,mean-std]).transpose()
    labels=['Mean+std','Mean','Mean-std']
    if ylabel is not None:
        axis.set_ylabel(ylabel)
    YP.plot(graphs,fig,axis,labels=labels,title='Mean+-std   '+m)
    
```

```python
plot_pair(['TMIN','TMAX'],plot_mean_std,ylabels=['Degree Celsius','Degree Celsius'],tenths=[True,True])
```

```python
plot_pair(['TOBS','PRCP'],plot_mean_std,ylabels=['Degree Celsius','Millimeter Per Day'],tenths=[True,True])
```

```python
plot_pair(['SNOW', 'SNWD'],plot_mean_std,ylabels=['Millimeter Per Day','Millimeter Accumulated'])
```

### plotting top 3 eigenvectors

```python
def plot_eigen(m,fig,axis,ylabel=None,tenth=None):
    EV=STAT[m]['eigvec']
    YP.plot(EV[:,:3],fig,axis,title='Top Eigenvectors '+m)
```

```python
plot_pair(['TMIN','TMAX'],plot_eigen)
```

```python
plot_pair(['TOBS','PRCP'],plot_eigen)
```

```python
plot_pair(['SNOW', 'SNWD'],plot_eigen)
```

### Script for plotting percentage of variance explained

```python
def pltVarExplained(j):
    subplot(1,3,j)
    EV=STAT[m]['eigval']
    k=5
    plot(([0,]+list(cumsum(EV[:k])))/sum(EV))
    title('Percentage of Variance Explained for '+ m)
    ylabel('Percentage of Variance')
    xlabel('# Eigenvector')
    grid()
    
```

```python
f=plt.figure(figsize=(15,4))
j=1
for m in ['TMIN', 'TOBS', 'TMAX']: #,
    pltVarExplained(j)
    j+=1
```
