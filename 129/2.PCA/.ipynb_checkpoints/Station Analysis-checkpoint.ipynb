```python
sqlContext.registerDataFrameAsTable(df,'weather')
Query="SELECT * FROM weather"
print Query
df1 = sqlContext.sql(Query)

def unpackRow(row):
    
    unpackedRow = dict()
    passThroughCols = ['elevation', 'latitude', 'longitude', 'measurement', 'station', 'undefs', 'year', 'label']
    for colName in passThroughCols:
        unpackedRow[colName] = row[colName]    
    unpackedRow['vector'] = unpackArray(row['vector'], np.float16)
    
    return unpackedRow

row_values = df1.rdd.map(unpackRow).collect()
X = pd.DataFrame(row_values)
```

### Patterns of how stations were present

```python
import seaborn as sns
sns.set_style('whitegrid')

station, counts = [], []
for g in X.groupby('station'):
    station.append(g[0])
    counts.append(len(g[1][g[1]['measurement'] == 'TOBS']))
    
#station_counts = [(g[0], len(g[1])) for g in X_post49.groupby('station')]
#station, counts = zip(*station_counts)
sns.distplot( 100* ( sorted(counts)/ ( ( max(year) - min(year) ) ) ) )
plt.xlabel('% of years with valid samples')
plt.ylabel('Density: ')
plt.xlim((0,100))
plt.show()
```

To see the impact, we measure the % of the year range that a station has a valid sample in terms of TOBS. We choose TOBS because we expect TOBS to have a valid sample when the station was present. This is in contrast to SNOW or PRCP etc which might have more variance.

We see that a large mode in % years close to 0%. Other modes are close to 20% and 50%. The hypothesis about 50% mode is that these are stations introduced post 1949 as per the pattern observed earlier. The higher percentage are observations for stations that were present throughout. We next analyze how the pattern of station prescense looks using a PCA analysis to try explaining all the modes and patterns observed here. 

```python
listIsValid = []
listUnSmoothed = []

def movingAvFilter(x, k = None):
    if k == None:
        k = int(len(x)/20)
    return np.array([np.mean(x[i-k:i+k]) for i in range(k+1,len(x)-k-1)])

for g in X.groupby('station'):
    isValid = np.zeros((int(max(year) - min(year) + 1),))
    posIdx = np.array(g[1]['year'].as_matrix(),int) - int(min(year))
    isValid[posIdx] = 1
    listUnSmoothed.append(isValid)
    listIsValid.append(movingAvFilter(isValid, 10))

unsmPrMatrix = np.array(listUnSmoothed)
prMatrix = np.array(listIsValid)
plt.plot(np.mean(prMatrix, axis = 1), marker = 'x')
print len(listIsValid)
```

```python
from sklearn.decomposition import PCA, DictionaryLearning
print prMatrix.shape

p = PCA(whiten=True)
p.fit(prMatrix)
plt.plot(range(1,11),1 - p.explained_variance_ratio_[:10], marker = 'o')
plt.ylim((0,1))
plt.xlabel('No of components')
plt.ylabel('Cum. % of variance explained')
plt.show()

plt.imshow(unsmPrMatrix[np.argsort(-np.mean(prMatrix, axis = 1)),:].T)
plt.show()
```
