### Distribution of missing observations
The distribution of missing observations is not uniform throughout the year. We visualize it below.

```python
def plot_pair(pair,func,myLabels=None,numPlots=2):
    j=0
    if numPlots==2:
        fig,X=subplots(1,2,figsize=(16,6))
        axes=X.reshape(2)
    else:
        fig,X=subplots(1,1,figsize=(16,6))
        axes=X
    #axes=X.reshape(2)
    axis=axes
    for m in pair:
        if numPlots==2:
            axis = axes[j]
        j+=1
        func(m,fig,axis,myLabels,numPlots)
        
def plot_valid(m,fig,axis,myLabels,numPlots=2):
    valid_m=STAT[m]['NE']
    if numPlots==2:
        YP.plot(valid_m,fig,axis,labels=myLabels,title='valid-counts '+m)
    else:
        YP.plot(valid_m,fig,labels=myLabels,title='valid-counts '+m)
    
```

```python
plot_pair(['TMIN','TMAX'],plot_valid,["sdas"])
```

```python
plot_pair(['TOBS','PRCP'],plot_valid)
```

```python
plot_pair(['SNOW', 'SNWD'],plot_valid)
```

### Plots of mean and std of observations

```python
def plot_mean_std(m,fig,axis,myLabels,numPlots):
    scale=1.0
    add=0
    if m in ['TMIN','TMAX']:
        scale=9.0/50
        add=32
    if m in ['PRCP']:
        scale=0.1
    mean=scale*STAT[m]['Mean']+add
    std=scale*np.sqrt(STAT[m]['Var'])+add
    myList=[mean-std,mean,mean+std]
    myList=[mean]
    print sum(mean)/len(mean)
    graphs=np.vstack(myList).transpose()
    axis.set_ylabel('Millimeter')
    #myLabels=['Mean+std','Mean','Mean-std']
    #axis.labels(labels=myLabels)
    #myLabels=['Mean '+m]
    YP.plot(graphs,fig,axis,labels=myLabels, title='Mean '+m)
```

```python
def plot_mean_std2(m,fig,axis,myLabels,numPlots):
    scale=9.0/50
    add=32
    myList=[]
    for m in ['TMIN','TMAX']:
        mean=scale*STAT[m]['Mean']+32
        std=scale*np.sqrt(STAT[m]['Var'])+32
    #myList=[mean-std,mean,mean+std]
        myList.append(mean)
    graphs=np.vstack(myList).transpose()
    #axis.set_ylabel('Fahrenheit')
    myLabels=['Mean TMIN','Mean TMAX']
    #axis.labels(labels=myLabels)
    #myLabels=['Mean '+m]
    YP.plot(graphs,fig,axis,labels=myLabels, title='Mean '+m)
```

```python
myLabels=['Mean TMIN','Mean TMAX','Mean+std']
myLabels=['Mean']
plot_pair(['TMIN'],plot_mean_std2,myLabels,numPlots=1)
```

```python
myLabels=['Mean-std','Mean','Mean+std']
myLabels=['Mean']
plot_pair(['PRCP'],plot_mean_std,myLabels,1)
```

```python
myLabels=['Mean']
plot_pair(['SNWD'],plot_mean_std,myLabels,numPlots=1)
#myLabels=["First Eigen Vector","Second Eigen Vector","Third Eigen Vector","Fourth Eigen Vector"]
N=3
myLabels=["eig"+str(i) for i in range(1,N+1)]
plot_pair(['SNWD'],plot_eigen,myLabels,numPlots=1)
```

```python
def plot_eigen(m,fig,axis,myLabels,numPlots):
    EV=STAT[m]['eigvec']
    #myLabels=["First Eigen Vector","Second Eigen Vector","Third Eigen Vector"]
    #axis.legend(labels=myLabels)
    YP.plot(EV[:,:3],fig,axis,labels=myLabels,title='Top Eigenvectors '+m)
```
