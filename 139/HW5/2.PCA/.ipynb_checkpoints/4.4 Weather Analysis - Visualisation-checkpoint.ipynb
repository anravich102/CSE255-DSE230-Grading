### Select data for a particular station and measurement type

```python
sqlContext.registerDataFrameAsTable(df,'weather')
Query="SELECT * FROM weather\n\tWHERE measurement='%s' and station='%s'"%('PRCP','USC00219059')
print Query
df1 = sqlContext.sql(Query)
print df1.count(),'rows'
df1.show(2)
rows=df1.rdd.map(lambda row:unpackArray(row['vector'],np.float16)).collect()
T=np.vstack(rows)
T=T/10.  # scaling to make the temperature be in centingrates
shape(T)
```

### Distribution of missing observations
The distribution of missing observations is not uniform throughout the year. We visualize it below.

```python
def plot_pair(pair,func):
    j=0
    fig,X=subplots(1,2,figsize=(16,6))
    axes=X.reshape(2)
    for m in pair:
        axis = axes[j]
        j+=1
        func(m,fig,axis)
#        func(m,fig,axis,['eig1','eig2','eig3'])
        
def plot_valid(m,fig,axis):
    valid_m=STAT[m]['NE']
    YP.plot(valid_m,fig,axis,title='valid-counts '+m)
    
```

```python
plot_pair(['TMIN','TMAX'],plot_valid)
```

```python
plot_pair(['TOBS','PRCP'],plot_valid)
```

```python
plot_pair(['SNOW', 'SNWD'],plot_valid)
```

### Plots of mean and std of observations

```python
def plot_mean_std(m,fig,axis):
    mean=STAT[m]['Mean']*0.1
    std=np.sqrt(STAT[m]['Var'])*0.1
    plt.ylabel('Temperature in Celsius')
    graphs=np.vstack([mean-std,mean,mean+std]).transpose()
#    graphs = mean.transpose()
    plt.ylabel('Mean Snowdepth')
    YP.plot(graphs,fig,axis,title='Mean+-std   '+m,labels=['mean-std','mean','mean+std'])
#    YP.plot(graphs,fig,axis,title='Mean   '+m,labels=['mean'])
```

```python
plot_pair(['TMIN','TMAX'],plot_mean_std)
```

```python
plot_pair(['TOBS','PRCP'],plot_mean_std)
```

```python
plot_pair(['SNOW', 'SNWD'],plot_mean_std)
```

### plotting top 3 eigenvectors

```python
def plot_eigen(m,fig,axis,lab=['eig1','eig2','eig3']):
    EV=STAT[m]['eigvec']
    YP.plot(EV[:,:3],fig,axis,title='Top Eigenvectors '+m,labels=lab)
```

```python
plot_pair(['TMIN','TMAX'],plot_eigen)
```

```python
plot_pair(['TOBS','PRCP'],plot_eigen)
```

```python
plot_pair(['SNOW', 'SNWD'],plot_eigen)
```

### Script for plotting percentage of variance explained

```python
def pltVarExplained(j):
    subplot(1,3,j)
    EV=STAT[m]['eigval']
    k=5
    plot(([0,]+list(cumsum(EV[:k])))/sum(EV))
    title('Percentage of Variance Explained for '+ m)
    ylabel('Percentage of Variance')
    xlabel('# Eigenvector')
    grid()
    
```

```python
f=plt.figure(figsize=(15,4))
j=1
for m in ['TMIN', 'TOBS', 'TMAX']: #,
    pltVarExplained(j)
    j+=1
```

```python
f=plt.figure(figsize=(15,4))
j=1
for m in ['SNOW', 'SNWD', 'PRCP']:
    pltVarExplained(j)
    j+=1 
```
